# Apagar Participantes Faltantes

```python
df = df.dropna(subset=['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_MT', 'NU_NOTA_LC', 'NU_NOTA_REDACAO'])
df.shape
```

# Adicionar Coluna Nota Total e Média

```python
df['W_Nota_Total'] = df['NU_NOTA_CN'] + df['NU_NOTA_CH'] + df['NU_NOTA_LC'] + df['NU_NOTA_MT'] + df['NU_NOTA_REDACAO']
df['W_Media'] = df['W_Nota_Total'] / 5
df.head()
```
