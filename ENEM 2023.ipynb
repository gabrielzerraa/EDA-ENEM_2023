# Apagar Participantes Faltantes

```python
df = df.dropna(subset=['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_MT', 'NU_NOTA_LC', 'NU_NOTA_REDACAO'])
```

```python
df.shape
```

# Adicionar Coluna Nota Total e Média

```python
df['W_Nota_Total'] = df['NU_NOTA_CN'] + df['NU_NOTA_CH'] + df['NU_NOTA_LC'] + df['NU_NOTA_MT'] + df['NU_NOTA_REDACAO']
df['W_Media'] = df['W_Nota_Total'] / 5
```

### Gênero

```python
# Tamanho do Gráfico
plt.figure(figsize=(9,6))

# Plotar Gráfico
ax = sns.countplot(df, x="TP_SEXO", color='#1e90ff')

# Adicionar Titulo
ax.set_title('Gênero - ENEM 2023', loc='left', fontsize=12)

# Remover a legenda dos Eixos
ax.set_xlabel('')
ax.set_ylabel('')

# Remover Gird
ax.spines[['right', 'left', 'top', 'bottom']].set_visible(False)

# Remover Legenda
plt.yticks([])
# Modificar Legenda
plt.xticks([0, 1], ['Mulher', 'Homem'])

# Total
total = df['TP_SEXO'].count()

# Total
total = df['TP_COR_RACA'].count()

# Adicionar anotações nas barras
ax.bar_label(ax.containers[0], fmt=lambda x: f'{(x/total)*100:0.1f}%')

# Adicionar margens
ax.margins(x=0.1, y=0.2)

# Visualizar o gráfico
plt.show()
```
