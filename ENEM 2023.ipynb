# Verificando Dados Nuloss

## Identificando Valores Ausentes

### Percentual dos Valores Ausentes

```python
# Total de Valores Ausentes
total_ausentes = df.isnull().sum().sum()

# Total de Células
total_celulas = np.product(df.shape)

# Percentual dos valores ausentes
percentual_ausentes = total_ausentes/total_celulas * 100

# Print
print(f'Porcentagem dos dados ausentes: {percentual_ausentes:.2f}%')
```

### Percentual de Valores Ausentes por Colunas

```python
# Total de Valores Ausentes
total_ausentes = df.isnull().sum() 

# Porcentagem de valores ausentes em cada coluna
percentual_por_coluna = total_ausentes / len(df) * 100

# Type de cada coluna
type_coluna = df.dtypes

# Criar um dataframe com os resultados
df_dados_linhas_ausentes = pd.DataFrame({
    'Valores Ausentes':total_ausentes,
    'Percentual':percentual_por_coluna,
    'Type':type_coluna
}).sort_values('Percentual', ascending=False)

df_dados_linhas_ausentes
```

# Visualização de Dados

```python
# Tamanho do Gráfico
plt.figure(figsize=(9,6))

ax = sns.countplot(df, x="TP_SEXO", color='#7F86D6')
#palette='Greys'

# Adicionar Titulo
ax.set_title('Gênero - ENEM 2023', loc='left', fontsize=12)

# Remover os Eixos
ax.set_xlabel('')
ax.set_ylabel('')

# Remover Gird
ax.spines[['right', 'left', 'top', 'bottom']].set_visible(False)

# Remover Legenda
plt.yticks([])
# Modificar Legenda
plt.xticks([0, 1], ['Homem', 'Mulher'])

# Total
total = df['TP_SEXO'].count()

# annotate the bars with fmt from matplotlib v3.7.0
ax.bar_label(ax.containers[0], fmt=lambda x: f'{(x/total)*100:0.1f}%')

# add space at the end of the bar for the labels
ax.margins(x=0.1)

# Inserir anotacoes
#plt.annotate('GENEROS QUE PREDOMINAM', xy=(0,0), xytext=(0,3))

plt.show()
```
