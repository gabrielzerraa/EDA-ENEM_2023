# Análise da Teoria de Resposta ao Item (TRI)

|O TRI é uma metodologia estatística utilizada no ENEM para calcular as notas de forma mais precisa e justa. O TRI leva em consideração a dificuldade da questão, a discriminação e a possibilidade de acerto ao acaso. 
Para considerar a dificuldade da questão, que pode ser fácil, média ou difícil, o TRI leva em conta que questões mais difíceis têm menos acertos e questões mais fáceis têm mais acerto 

A discriminação é a capacidade da questão de diferenciar o nível de habilidade do candido.

A possibilidade de acerto ao acaso estima a chance de o estudante acertar a questão chu  
ndo.

Será feita uma análise mostrando a nota mínima, média e máxima para uma mesma quantidade de acertos. Além disso, será apresentada a diferença de pontos entre a maior e a menor nota, bem como o percentual que a maior nota supera   a menor.

```python
def qtde_questoes_acertadas(str1, str2):
    tamanho = len(str1)
    contador = 0
    for i in range(tamanho):
        if str1[i] == str2[i]:
            contador += 1
    return contador
```
