# DEPOIS MUDAR

```python
def qtde_questoes_acertadas(str1, str2):
    tamanho = len(str1)
    contador = 0
    for i in range(tamanho):
        if str1[i] == str2[i]:
            contador += 1
    return contador
```

## Ciências da Natureza

```python
df_notas_cn = df[['TX_GABARITO_CN', 'TX_RESPOSTAS_CN', 'NU_NOTA_CN']]
df_notas_cn['Qtde_Questoes'] = df_notas_cn.apply(lambda row: qtde_questoes_acertadas(row['TX_GABARITO_CN'], row['TX_RESPOSTAS_CN']), axis=1)

minima = df_notas_cn.groupby('Qtde_Questoes')['NU_NOTA_CN'].min()
media = df_notas_cn.groupby('Qtde_Questoes')['NU_NOTA_CN'].mean()
maxima = df_notas_cn.groupby('Qtde_Questoes')['NU_NOTA_CN'].mean()
diferenca = maxima - minima

notas_cn = {'Minima' : minima,
        'Média' :  media,
        'Máxima' : maxima,
        'Diferenca' : diferenca}

pd.DataFrame(notas_cn)
```
